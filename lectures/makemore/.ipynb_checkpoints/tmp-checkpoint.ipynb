```python

import matplotlib.pyplot as plt
%matplotlib inline
#  N_tri ( [27, 27, 27]  PyTorch Tensor)
#  itos 
#  stoi 

def visualize_trigram_counts(prefix):
    """
     (prefix) 
    """
    # 1. 
    if len(prefix) != 2:
        print(" 2 ")
        return
    
    # 2. 
    ix1 = stoi.get(prefix[0])
    ix2 = stoi.get(prefix[1])
    
    if ix1 is None or ix2 is None:
        print(f" '{prefix}' ")
        return

    # 3.  3D 
    #  N[ix1, ix2, :]
    counts_vector = N[ix1, ix2, :]
    
    # 4.  0
    if counts_vector.sum() == 0:
        print(f" '{prefix}' ")

    # 5. 
    # x  ('.', 'a', 'b', ..., 'z')
    labels = [itos[i] for i in range(27)]
    
    plt.figure(figsize=(15, 6)) # 
    plt.bar(labels, counts_vector.numpy()) # 
    
    # 6. 
    plt.title(f"Trigram Counts: Given prefix '{prefix}', what comes next?")
    plt.xlabel("Next Character")
    plt.ylabel("Frequency (Count)")
    plt.show()

# --- ---

#  'an' 
visualize_trigram_counts('an')

#  'jo' 
visualize_trigram_counts('jo')

#  'th' 
visualize_trigram_counts('th')

#  'qz'
visualize_trigram_counts('qz')
```

```python
import torch
import torch.nn.functional as F
import matplotlib.pyplot as plt # ()
%matplotlib inline
```

```python
# --- 1.  ---
words = open('names.txt', 'r').read().splitlines()
chars = sorted(list(set(''.join(words))))
stoi = {s:i+1 for i,s in enumerate(chars)}
stoi['.'] = 0
itos = {i:s for s,i in stoi.items()}
vocab_size = 27

# --- 2.  (Trigram) ---
xs, ys = [], []
print("...")

for w in words:
    #  ['.', '.'] 
    chs = ['.', '.'] + list(w) + ['.']
    for ch1, ch2, ch3 in zip(chs, chs[1:], chs[2:]):
        ix1 = stoi[ch1]
        ix2 = stoi[ch2]
        ix3 = stoi[ch3]
        
        # xs  ** 
        xs.append([ix1, ix2])
        
        # ys  ** 
        ys.append(ix3)

xs = torch.tensor(xs) # : [196113, 2]
ys = torch.tensor(ys) # : [196113]
num = xs.shape[0]
print(f" () : {num}")
print(f"xs : {xs.shape}")
print(f"ys : {ys.shape}")
```

```python
# --- 3.  "W" ---
g = torch.Generator().manual_seed(2147483647)

# 
#  = 27 ( ch1) + 27 ( ch2) = 54
#  = 27 ( ch3  logits)
W = torch.randn((54, 27), generator=g, requires_grad=True)
print(f"W () : {W.shape}")
```

```python
# --- 4.  ---

# 
learning_rate = 1.0 #  ()
num_epochs = 100000     #  ()
batch_size = 32      #  ()

print("\n---  ---")

for i in range(num_epochs):
    
    # ---  (mini-batch) ---
    ix = torch.randint(0, num, (batch_size,), generator=g)
    xs_batch = xs[ix] # : [32, 2]
    ys_batch = ys[ix] # : [32]
    
    # ---  (Forward Pass) ---
    
    # 1.  (One-hot)
    #  **  (ch1)
    xenc1 = F.one_hot(xs_batch[:, 0], num_classes=vocab_size).float() # : [32, 27]
    #  **  (ch2)
    xenc2 = F.one_hot(xs_batch[:, 1], num_classes=vocab_size).float() # : [32, 27]
    
    # 2.  (Concatenate)
    #  [32, 27]  [32, 54] 
    xenc = torch.cat([xenc1, xenc2], dim=1) 
    
    # 3.  Logits ( Bigram )
    # [32, 54] @ [54, 27] --> [32, 27]
    logits = xenc @ W 
    
    # 4.  (Loss)
    # F.cross_entropy  softmax  NLL Loss
    # 
    loss = F.cross_entropy(logits, ys_batch)
    
    # ---  (Backward Pass) ---
    W.grad = None # 
    loss.backward() # 
    
    # ---  (Update) ---
    W.data += -learning_rate * W.grad
    
    # ---  ---
    if i % (num_epochs // 10) == 0 or i == num_epochs - 1:
        print(f'Epoch {i:4d}/{num_epochs} | Loss: {loss.item():.4f}')

print("---  ---")
```

```python
# --- 5.  Loss ---

#  ()
with torch.no_grad(): # 
    
    # 1.  xs
    xenc1_full = F.one_hot(xs[:, 0], num_classes=vocab_size).float() #  [196113, 27]
    xenc2_full = F.one_hot(xs[:, 1], num_classes=vocab_size).float() #  [196113, 27]
    
    # 2. 
    xenc_full = torch.cat([xenc1_full, xenc2_full], dim=1) #  [196113, 54]
    
    # 3.  Logits
    logits_full = xenc_full @ W #  [196113, 27]
    
    # 4.  Loss
    full_loss = F.cross_entropy(logits_full, ys)
    
    print("\n---  ---")
    print(f' (Bigram) Loss ():      ~2.45')
    print(f' (Trigram) Loss ():    ~{nll.item():.4f}') #  nll
    print(f' (Trigram) Loss (): {full_loss.item():.4f}')

#  Loss  Loss (2.20) 
```

```python
# --- 6.  (Sampling) ---
print("\n---  () ---")
g_sample = torch.Generator().manual_seed(2147483647 + 10) # 

for _ in range(10):
    
    out = []
    # 1. 
    context1 = 0 # '.'
    context2 = 0 # '.'
    
    while True:
        # ---  () ---
        # 1.  (: batch_size=1)
        xenc1 = F.one_hot(torch.tensor([context1]), num_classes=vocab_size).float() # [1, 27]
        xenc2 = F.one_hot(torch.tensor([context2]), num_classes=vocab_size).float() # [1, 27]
        # 2. 
        xenc = torch.cat([xenc1, xenc2], dim=1) # [1, 54]
        
        # ---  (Inference) ---
        # 3.  Logits
        logits = xenc @ W # [1, 54] @ [54, 27] --> [1, 27]
        # 4. 
        probs = F.softmax(logits, dim=1)
        
        # 5. 
        ix = torch.multinomial(probs, num_samples=1, generator=g_sample).item()
        
        # 6. 
        if ix == 0:
            break
        
        out.append(itos[ix])
        
        # 7. "" 
        context1 = context2
        context2 = ix
        
    print(''.join(out))
```
